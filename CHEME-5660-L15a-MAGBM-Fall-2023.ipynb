# Example: Multidimensional Geometric Brownian Motion (MGBM)
Consider an asset portfolio $\mathcal{P}$ with a return covariance matrix $\Sigma$ and drift vector $\mu$. The multi-dimensional geometric Brownian motion model describing the share price $S_{i}(t)$ for asset $i\in\mathcal{P}$ is given by: 

$$
\begin{equation*}
\frac{dS_{i}\left(t\right)}{S_{i}(t)} = \mu_{i}\,{dt}+\sum_{j\in\mathcal{P}}a_{ij}\cdot{dW_{j}(t)}\qquad\text{for}\quad{i \in \mathcal{P}}
\end{equation*}
$$

where $a_{ij}\in\mathbf{A}$ and $\mathbf{A}\mathbf{A}^{\top} = \Sigma$. The multi-dimensional GBM model has the analytical solution:

$$
\begin{equation*}
S_{i}(t) = S_{i,\circ}\cdot\exp\Biggl[\left(\mu_{i}-\frac{\sigma_{i}^{2}}{2}\right)\Delta{t} + \sqrt{\Delta{t}}\cdot\sum_{j\in\mathcal{P}}a_{ij}\cdot{Z_{j}(0,1)}\Biggr]\quad{i\in\mathcal{P}}
\end{equation*}
$$

where $S_{i,\circ}$ is the initial share price at $t_{\circ}$ for asset $i\in\mathcal{P}$,  $\Delta{t} = t - t_{\circ}$ is the time difference, and $Z_{j}(0,1)$ is a standard normal random variable.

## Learning objective
This example aims to familiarize students with simulating the multi-dimensional geometric Brownian motion (MGBM) using a factorized version of the covariance matrix. This approach was reproduced from Chapter 3 of Glasserman:

* [Amazon: Monte Carlo Methods in Financial Engineering, Glasserman, Springer-Verlag, 2003, ISBN 10: 0387004513](https://www.amazon.com/Financial-Engineering-Stochastic-Modelling-Probability/dp/0387004513)  

## Setup

```julia
include("Include.jl");
```

### Set some constants for the simulation

```julia
Δt = (1.0/252.0);
T = 48.0*(Δt);
number_of_steps = 48;
number_of_states = 2;
number_of_trials = 1;
```

## Setup the model
Let's set some values for the drift vector $\mu$, and the covariance matrix $\Sigma$. For now we'll just pick illsutrative values, but for an application we would compute these values from historical data:

```julia
μ = [0.10,0.05];
```

```julia
Σ = zeros(2,2);
Σ[1,1] = 0.10
Σ[1,2] = -0.05
Σ[2,1] = -0.05
Σ[2,2] = 0.05;
```

Glasserman factors the covariance matrix $\Sigma = \mathbf{A}\mathbf{A}^{\top}$ using a [Cholesky factorization](https://en.wikipedia.org/wiki/Cholesky_decomposition). This factorization is implemented in the [LinearAlgebra.jl](https://docs.julialang.org/en/v1/stdlib/LinearAlgebra/):

```julia
factors = cholesky(Σ);
A = factors.L |> x-> Matrix(x)
Â = diagm(0 => diag(A))
```

#### Check: Is Cholesky legit?

```julia
@assert isapprox(Σ, A*transpose(A),rtol = 1e-1)
```

```julia
A*transpose(A)
```

### Set the initial conditions and the time range

```julia
Sₒ = [20.0,20.0];
time_array = range(0, stop = T, length = number_of_steps) |> collect;
```

## MGBM Simulation

```julia
simulation_dictionary = Dict{Int64,Array{Float64,2}}();
for trial_index ∈ 1:number_of_trials
    simulation_array = Array{Float64,2}(undef, number_of_steps, 3);
    for i ∈ eachindex(time_array)
        t = time_array[i];
        simulation_array[i,1] = t;

        for j ∈ 1:number_of_states
            simulation_array[i,j+1] = Sₒ[j]*exp((μ[j] - Σ[j,j]/2)*t + (√t)*dot(A[j,:],randn(number_of_states)));
        end
    end
    simulation_dictionary[trial_index] = simulation_array;
end
```

```julia
p = plot();
for (key, simulation_array) ∈ simulation_dictionary
    
    plot!(simulation_array[:,1],simulation_array[:,2], label="1", c=:gray69);
    plot!(simulation_array[:,1],simulation_array[:,3], label="2", c=:red)
end
current()
xlabel!("Time (years)", fontsize=18)
ylabel!("Share price (USD/share)", fontsize=18)
```

### Check the correlation

```julia
x = simulation_dictionary[1][:,2];
y = simulation_dictionary[1][:,3];
```

```julia
cor(x,y)
```
